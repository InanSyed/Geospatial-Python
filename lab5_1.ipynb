```python
#import libraries
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt
import folium


##load data from: https://opencontext.org/projects/95b1ef01-9ccb-4484-b0f2-540f4dc54672
mosques = pd.read_csv("mosques.csv")

#make into geodataframe
geometry =[Point(xy) for xy in zip(mosques["Longitude (WGS-84)"], mosques["Latitude (WGS-84)"])]
mosques= gpd.GeoDataFrame(mosques, geometry=geometry, crs="EPSG:4326")


#making an interactive map using folium
lon = mosques.geometry.x.mean()
lat = mosques.geometry.y.mean()
map = folium.Map(location=[lat, lon])

#markers for each mosque
for i, row in mosques.iterrows():
    popup_content = f"""
    <b>{row['Mosque name [https://opencontext.org/predicates/c7f34365-cfcf-46dd-856f-7bef8d8a8421]']}</b><br>
    <i>Location:</i> {row['City/site [https://opencontext.org/predicates/04c1f98d-a040-4914-99d5-6fc4711f7062]']}, {row['State [https://opencontext.org/predicates/9d10707f-3642-4616-ace1-5b498e2845f0]']}<br>
    <i>Date:</i> {row['Concluded date for the mosque [https://opencontext.org/predicates/b853c72e-973c-45e5-b65d-603fc4aa4ed9]']}<br>
    <i>Form:</i> {row['Form of the mosque: square, rectangular with longitudinal direction (basilica), rectangular with transverse direction, other [https://opencontext.org/predicates/9cbca203-081f-44ed-9b3e-fd8226c091b0]']}<br>
    <i>Mihrab:</i> {row['A mihrab is present [https://opencontext.org/predicates/831a6a82-9e56-494e-810d-6e047e08b5ce]']}<br>
    <i>Construction materials:</i> {row['Construction materials [https://opencontext.org/predicates/8f0d88d1-6b50-4ede-ad1d-b554313b4364]']}    """

    folium.Marker(
        location=[row["Latitude (WGS-84)"], row["Longitude (WGS-84)"]],
        popup=folium.Popup(popup_content, max_width=300),
        icon=folium.Icon(color="black", icon="info-sign")
    ).add_to(map)


#show it
map
```
